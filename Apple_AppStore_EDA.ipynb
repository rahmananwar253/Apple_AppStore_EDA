### 4.5 Handling missing data
- In the `Developer_website` column, we have 52% missing values.So, instead of imputing those missing values, we should remove this column.
- if the percentage of missing values is high and you believe that the column will not provide meaningful insights, removing it is also an option.

```python
# drop the developer_Website column as it is not required for our analysis
df.drop('Developer_Website', axis=1, inplace=True)
```

```python
# Replace null values in `Developer_Url` column with "No_Url"
df['Developer_Url'].fillna('No_Url', inplace=True)
```

```python
# Now we have to check missing values again
df.isnull().sum().sort_values(ascending=False).head()
```
